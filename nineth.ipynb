```python
feats = pd.read_csv(
    'Applied-Deep-Learning-with-Keras/' +
    'Lesson01/data/bank_data_feats_e3.csv',
    index_col = 0
)
target = pd.read_csv(
    'Applied-Deep-Learning-with-Keras/' +
    'Lesson01/data/bank_data_target_e2.csv',
    index_col = 0
)
```

```python
from sklearn.model_selection import train_test_split

test_size = 0.2
random_state = 42
# y = Mx + b
# 학습시킬 정보, 테스트할 정보를 적절하게 만든다.
# test_size 학습에 적용하는 비율을 결정한다.
# test_size = 0.2
# 학습에 80%의 데이터를 활용하고
# 테스트에 20%의 데이터를 사용한다는 의미
# random_state는 재현율을 지정해주는 부분이라고 보면 된다.
X_train, X_test, y_train, y_test = train_test_split(
    feats, target, 
    test_size = test_size,
    random_state = random_state
)
```

```python
print(f'Shape of X_train: {X_train.shape}')
print(f'Shape of y_train: {y_train.shape}')
print(f'Shape of X_test: {X_test.shape}')
print(f'Shape of y_test: {y_test.shape}')
```

```python
from sklearn.linear_model import LogisticRegression

# y = 1 / (1 + e^-x)
model = LogisticRegression(random_state = 42)
```

```python
model.fit(X_train, y_train['y'])
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn import metrics

accuracy = metrics.accuracy_score(
    y_pred = y_pred, y_true = y_test
)

print(f'Accuracy of the model is {accuracy * 100:.4f}%')
```
